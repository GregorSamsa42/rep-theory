```python
def mult_trivial_rep(G, chi):
    """
    Find multiplicity of 1 in chi, which is a character of G, assumed to be in the order of G.conjugacy_classes()
    
    Parameters:
    -----------
    G group, chi character of G
    
    Returns:
    --------
    integer
    """
    classes = list(G.conjugacy_classes())
    class_sizes = [G.order() / G.centralizer(C.representative()).order() for C in classes]
    result = 0
    for idx in range(0,len(chi)):
        result += class_sizes[idx]*chi[idx]
    return result/G.order()
    
```

```python
def tensor_rep(chi1,chi2):
    if (len(chi1) != len(chi2)):
        raise ValueError("characters must have the same length")
    return [chi1[i]*chi2[i] for i in range(0, len(chi1))]
```

```python
n = 7
G = SymmetricGroup(n)
chartable = G.character_table()
irreps_amount = chartable.nrows()
for irrep1 in range(0, irreps_amount):
    for irrep2 in range(0, irreps_amount):
        # compute the tensor product of irrep1, irrep2 and then compute how often the trivial rep occurs
        # ignore the trivial rep times anything as this definitely gives 
        print(irrep1, irrep2, mult_trivial_rep(G, tensor_rep(chartable[irrep1],chartable[irrep2])))
```
